subjects as axis and marker as y axis

```python
stu1=[56,78,92,95,78]
stu2=[77,89,91,23,45]
sub=["Maths","Science","English","Telugu","Social"]
plt.plot(sub,stu1,label="student1")
plt.plot(sub,stu2,label="student2")
plt.title("marks vs subjects")
plt.legend()
```

```python
stu1=[56,78,92,95,78]
stu2=[77,89,91,23,45]
sub=["Maths","Science","English","Telugu","Social"]
plt.subplot(2,1,1)
plt.plot(sub,stu1,label="student1",color="g")
plt.legend()
plt.subplot(2,1,2)
plt.plot(sub,stu2,label="student2",color="y")
plt.legend()
```

```python
A=np.array([230,560,780,127,128])
B=np.array([200,160,270,127,400])
print("Revenues of company A ",A)
print("Revenues of company B ",B)
profA=np.diff(A)
print("profits of A ",profA)
profB=np.diff(B)
print("profits of B ",profB)
x=["19-20","20-21","21-22","22-23"]
plt.subplot(1,2,1)
plt.bar(x,profA,label="profit A",color='k')
plt.legend(loc = "best")
plt.subplot(1,2,2)
plt.bar(x,profB,label="profit B",color='g')
plt.legend(loc = "best")
```
